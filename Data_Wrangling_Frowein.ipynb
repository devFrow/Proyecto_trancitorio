```python
#importe de librerias y lectura de archivos 

import pandas as pd 
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt 
import seaborn as sns 
from sklearn.linear_model import LinearRegression
import statsmodels.formula.api as sm

url= 'https://raw.githubusercontent.com//devFrow//Proyecto_coder//main//train.csv'

df = pd.read_csv(url, sep = ',')

pd.set_option('display.max_rows', None)
pd.set_option('display.max_column', None)

print(df.shape)
df.head()
```

```python
#Exploracion de la variable objetivo

print(df['SalePrice'].describe().round(0))
promedio = df['SalePrice'].mean()
mediana = df['SalePrice'].median()
#vemos su distribución

plt.figure(figsize=(7,3))
plt.hist(df['SalePrice'], bins=30, color='blue', edgecolor='k', alpha=0.5)
plt.title('Histograma de SalePrice')
plt.xlabel('SalePrice')
plt.ylabel('Frecuencia')
plt.xticks(fontsize=8)
plt.axvline(promedio, color='purple', linewidth=1.5, label = f'Promedio : {promedio:.2f}')
plt.axvline(mediana, color='cyan', linewidth=1.5, label = f'Mediana : {mediana:.2f}')
plt.legend()
plt.show()
```

# Tratamiento de nulos 

vamos a ver caso por caso los valores nulos o nan para ver de que manera podemos tratarlos 

```python
#Tratamiento de nulos

#PoolQC
#Hace referencia a el estado de las piletas

# Ex - Excellent
# Gd - Good
# TA - Average/Tipical 
# Fa - Fair
# Po - Poor
# Na - No Pool ---> los na son un dato en si el hecho de no tener pileta

df['PoolQC'].fillna('None', inplace=True)


#MiscFeature
# Características varias no incluidas en otras categorías

# Elev	Elevator
# Gar2	2nd Garage (if not described in garage section)
# Othr	Other
# Shed	Shed (over 100 SF)
# TenC	Tennis Court
# NA	None  ---> los valores con na vuelven a significar que no posee ninguna caracteristica 

df['MiscFeature'].fillna('None', inplace=True)

#remplasamos vacios con none para indicar que no poseen carateristicas especiales .

#Alley
#hace referencia a tipo de acceso secundario o tracero 

#Grvl - Graved
#Pave - Paved
#Na - No alley access - cambiaremos los na por None 

df['Alley'].fillna('None', inplace=True)

#Fence 
#hace referencia a la calidad de la valla o cerca

#GdPrv - Good Privacy
#MnPrv - Minimum Privacy
#GdWo -	 Good Wood
#MnWw -	 Minimum Wood/Wire
#NA	  -  No Fence -> cambiamos por none 

df['Fence'].fillna('None', inplace=True)

#FireplaceQu
#Hace referencia a la calidad de la chimenea 

#Ex 	Excellent - Exceptional Masonry Fireplace
#Gd	    Good - Masonry Fireplace in main level
#TA	    Average - Prefabricated Fireplace in main living area or Masonry Fireplace in basement
#Fa 	Fair - Prefabricated Fireplace in basement
#Po	    Poor - Ben Franklin Stove
#NA	    No Fireplace  -> None

df['FireplaceQu'].fillna('None', inplace=True)

#LotFrontage 
#Proporciona información sobre la longitud, en pies (lineales), de la calle o vía de acceso que está directamente conectada a la propiedad

#como es un dato numerico que esta faltando  voy a remplazarlo con la media de las casas del mismo barrio 

med_lot = df.groupby('Neighborhood')['LotFrontage'].median()

df['LotFrontage'] = df.apply(lambda row: med_lot[row['Neighborhood']] if pd.isna(row['LotFrontage']) else row['LotFrontage'], axis=1)




```

# Tratamiento de categoricas.

Avancemos primero con las que procesamos con el propósito de limpiar los Na.

**PoolQC**

Esta variable habla de la calidad de la pileta en caso desde lo mas bajo (no tener) a lo mas alto (exelente)
por lo tanto es una categoria ordinal y la podemos factorizar. 


```python
#creo un diccionario para usar de mapeo
fact_Qc = {"Ex" : 5,"Gd" : 4,"TA" : 3,"Fa" : 2,"Po" : 1,"None" : 0}
df["PoolQC"] = df['PoolQC'].map(fact_Qc)
```

### -------------------------------------------------------------------------------------------

Calcualo la matrix de correlacion para hacer foco primero en las variables mas interesantes .


```python
mx_core = df.corr(method='pearson')
mx_t_core = mx_core['SalePrice']
mx_t_core = abs(mx_t_core)
mx_t_core = mx_t_core.drop(columns = 'SalePrice', index = 'SalePrice')
mx_t_core = mx_t_core.sort_values(ascending = False )
```

vamos a graficar para tener una vista clara del resultado 

```python
top_15 = mx_t_core[0:15]
plt.figure(figsize=(7,3))
plt.barh(top_15.index ,top_15.values,color='blue', edgecolor='k', alpha=0.5)
plt.gca().invert_yaxis()
plt.title('Tabla de Correlaciones top 15')
plt.xlabel('Correlacion con SalePrice')
plt.yticks(fontsize=8)

```

hacemos regresion para una primera vista

```python
model_r_1 = 'SalePrice ~ OverallQual + GrLivArea + GarageCars + GarageArea + TotalBsmtSF'

rl_m1  = sm.ols(formula=model_r_1, data = df ).fit()

print(rl_m1.summary())
```
